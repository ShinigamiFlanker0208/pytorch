```python
# comparing the time of the cpu_exec and gpu_exec
import time
import torch

size=10000
device=torch.device('cuda')

matrix1=torch.randn(size,size)
matrix2=torch.randn(size,size)

start_timecpu= time.time()
result_cpu=torch.matmul(matrix1,matrix2)
end_timecpu= time.time() - start_timecpu
print(f"time on CPU: {end_timecpu:.2f} seconds")

# now running it into GPU

mat1=matrix1.to('cuda')
mat2=matrix2.to('cuda')

start_timegpu=time.time()
result_gpu=torch.matmul(mat1,mat2)
torch.cuda.synchronize()
end_timegpu=time.time() - start_timegpu


print(f"time on GPU: {end_timegpu:.2f} seconds")


# compare results
print('\n speedup cpu time/gpu time',end_timecpu/end_timegpu)
```
